```python
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
X = np.array([[1, 20],
              [2, 21],
              [3, 22],
              [6, 25],
              [7, 26],
              [8, 27]])
```

```python
y = np.array([1, 0, 0, 1, 0, 1])
model = GaussianNB()
model.fit(X, y)
test = np.array([[4, 23]])
prediction = model.predict(test)
print("Predicted Class:", prediction[0])
```
